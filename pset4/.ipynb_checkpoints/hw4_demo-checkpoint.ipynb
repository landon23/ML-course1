1. In this homework set, we implement linear regression and compare it with various families of basis functions.

1.a) Fit a linear regression model using up to order $L=5$ monomials.  Plot the test data points and predictions for this model. 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import loadmat
data = loadmat('motor.mat')
import hw4
import imp
imp.reload(hw4)
motorReg = hw4.linReg()
motorReg.loadData(data, list(data.keys())[3:7])
motorReg.polyLinReg(5)
motorReg.fit()
y5 = motorReg.predict(motorReg.xTest)
motorReg.plot()
```

1.b) Fit models with polynomial features up to order $L=19$.  Plot the MSE on the training and test data on the same set of axes as a function of $L$.  Which model has the smallest training error and which has the smallest test error?  Plot the predictions of these models for the test data on the same axes as those from (a).

```python
[trainE, testE] = motorReg.errorRange(19)
plt.scatter(np.arange(19)+1, trainE, label='training error')
plt.scatter(np.arange(19)+1, testE, label='test error')
plt.ylabel('MSE')
plt.xlabel('Degree of polynomial feature')
plt.title('Training and test error')
plt.legend()
plt.show()

lTest = np.argmin(testE)+1
lTrain = np.argmin(trainE)+1

print('The model with the smallest training error has degree', lTrain, 'polynomials while the model with the smallest test error has polynomials of degree', lTest)

motorReg.polyLinReg(lTest)
motorReg.fit()
testPredict = motorReg.predict(motorReg.xTest)

motorReg.polyLinReg(lTrain)
motorReg.fit()
trainPredict = motorReg.predict(motorReg.xTest)


plt.scatter(motorReg.featTest, testPredict, label='best test error')
plt.scatter(motorReg.featTest, y5, label='degree 5 polynomial')
plt.scatter(motorReg.featTest, trainPredict, label='best train error')
plt.scatter(motorReg.featTest, motorReg.yTest, label='test data')
plt.legend()
plt.show()
```

1c) Consider instead of polynomial features, radial basis functions:
$$
\phi_j (x) = \exp \left[ - \frac{ (x - \mu_j )^2}{2 \sigma^2} \right]
$$
where $\mu_j$ are $L$ equally spaced points between $-1$ and $1$, and $\sigma^2 = ( \mu_2 - \mu_1)^2$ the spacing between consecutive centers.  Repeat parts (a)-(b) for $L=2, \dots, 25$.

```python
motorReg.radialLinReg(5)
motorReg.fit()
y5 = motorReg.predict(motorReg.xTest)
motorReg.plot()

[trainE, testE] = motorReg.radialRange(25)
plt.scatter(np.arange(24)+2, trainE, label='training error')
plt.scatter(np.arange(24)+2, testE, label='test error')
plt.ylabel('MSE')
plt.xlabel('Number of RBF')
plt.title('Training and test error')
plt.legend()
plt.show()

lTest = np.argmin(testE)+2
lTrain = np.argmin(trainE)+2

print('The model with the smallest training error has degree', lTrain, 'polynomials while the model with the smallest test error has polynomials of degree', lTest)

motorReg.radialLinReg(lTest)
motorReg.fit()
testPredict = motorReg.predict(motorReg.xTest)

motorReg.radialLinReg(lTrain)
motorReg.fit()
trainPredict = motorReg.predict(motorReg.xTest)


plt.scatter(motorReg.featTest, testPredict, label='best test error')
plt.scatter(motorReg.featTest, y5, label='5 RBF')
plt.scatter(motorReg.featTest, trainPredict, label='best train error')
plt.scatter(motorReg.featTest, motorReg.yTest, label='test data')
plt.legend()
plt.show()
```

2.a, b) Consider a Bayesian setting in which the coefficients $w$ are assigned a centered Gaussian prior with covariance $\alpha^{-1} \mathbb{1}$.  In this case the MAP estimate for $w$ is 
$$
w = ( \Phi^T \Phi +  \alpha \beta^{-1} \mathbb{1} )^{-1} \Phi^T y,
$$
where $\Phi$ is the data matrix and $\beta$ is the parameter in the assumption $y \vert_{x, w} \sim \mathrm{Normal} (w^T x), \beta^{-1} )$.  

With $L=50$ polynomial basis features, compute MAP estimates for $\beta = 0.0025$ and $ \alpha \in \{ 10^{-8}, 10^{-7}, \dots , 10^0 \}$.  Plot the training and test MSE as a function of $\log ( \delta)$.  Plot the predictions of the models which minimize test and training error.

```python
deltasPoly = np.power(10.0, np.arange(-8, 1)) / 0.0025
L = 50
[trainE, testE] = motorReg.errorRegRange(deltasPoly, L, basis='poly')
plt.scatter(-np.arange(-8, 1), trainE, label='training error')
plt.scatter(-np.arange(-8, 1), testE, label='test error')
plt.xlabel('- log(alpha)')
plt.ylabel('MSE')
plt.title('Training and test error for different regularizations')
plt.legend()
plt.show()

[iTrainPoly, iTestPoly] = [np.argmin(trainE), np.argmin(testE)]
motorReg.fitReg(deltasPoly[iTrainPoly])
trainP = motorReg.predict(motorReg.xTest)
motorReg.fitReg(deltasPoly[iTestPoly])
testP = motorReg.predict(motorReg.xTest)

plt.scatter(motorReg.featTest, trainP, label='best training error')
plt.scatter(motorReg.featTest, testP, label='best test error')
plt.scatter(motorReg.featTest, motorReg.yTest, label='test data')
plt.title('Test data vs predictions')
plt.legend()
plt.show()
```

2c) Repeat the previous problem with $L=50$ radial basis functions.

```python
deltasRadial =  np.power(10.0, np.arange(-8, 8)) /0.0025

[trainE, testE] = motorReg.errorRegRange(deltasRadial, L, basis='radial')
plt.scatter(-np.arange(-8, 8), np.log(trainE), label='training error')
plt.scatter(-np.arange(-8, 8), np.log(testE), label='test error')
plt.xlabel('- log(alpha)')
plt.ylabel('log(MSE)')
plt.title('Training and test error for different regularizations')
plt.legend()
plt.show()

[iTrainRadial, iTestRadial] = [np.argmin(trainE), np.argmin(testE)]
motorReg.fitReg(deltasRadial[iTrainRadial])
trainP = motorReg.predict(motorReg.xTest)
motorReg.fitReg(deltasRadial[iTestRadial])
testP = motorReg.predict(motorReg.xTest)

#plt.scatter(motorReg.featTest, trainP, label='best training error')
plt.scatter(motorReg.featTest, testP, label='best test error')
plt.scatter(motorReg.featTest, motorReg.yTest, label='test data')
plt.title('Test data vs predictions (best test error)')
plt.legend()
plt.show()

plt.scatter(motorReg.featTest, trainP, label='best training error')
plt.scatter(motorReg.featTest, motorReg.yTest, label='test data')
plt.title('Test data vs predictions (best training error)')
plt.legend()
plt.show()
```

2.d) The posterior mean for $w$ is given in the problem statement for 2a above.  The posterior covariance is,
$$
S = ( \beta \Phi^T \Phi + \alpha \mathbb{1} )^{-1}
$$
Draw 10 samples for $w$ and plot the resulting prediction functions $f(x) = w^T \Phi (x)$, for each of the models minimizing the test error for the $50$ polynomial and radial basis function examples above.

```python
motorReg.polyLinReg(50)
motorReg.fitReg(deltasPoly[iTestPoly])
m = motorReg.w
#print(motorReg.w.shape)
sInv = 0.0025* np.matmul(motorReg.phi.transpose(), motorReg.phi)+ (deltasPoly[iTestPoly] *0.0025) * np.identity(motorReg.phi.shape[1])
w = np.random.multivariate_normal(m.flatten(), (np.linalg.inv(sInv)+ np.linalg.inv(sInv).transpose())/2, 10)
#w = np.random.multivariate_normal(m.flatten(), 0.0000000000000000000000000000000001* np.identity(len(m)), 10)
#plt.scatter(motorReg.featTest, motorReg.predict(motorReg.xTest))
for i in range(10):
    motorReg.w = np.reshape(w[i, :], (-1,1))
    #print(motorReg.w.shape)
    plt.plot(motorReg.featGrid, motorReg.predict(motorReg.xGrid))
plt.scatter(motorReg.featTest, motorReg.yTest, label='test data')
plt.legend()
plt.title('Random instances of prediction function for polynomial features')
plt.show()

motorReg.radialLinReg(50)
motorReg.fitReg(deltasRadial[iTestRadial])
m = motorReg.w
#print(motorReg.w.shape)
sInv = 0.0025* np.matmul(motorReg.phi.transpose(), motorReg.phi)+ (deltasRadial[iTestRadial] *0.0025) * np.identity(motorReg.phi.shape[1])
w = np.random.multivariate_normal(m.flatten(), (np.linalg.inv(sInv)+ np.linalg.inv(sInv).transpose())/2, 10)
#w = np.random.multivariate_normal(m.flatten(), 0.0000000000000000000000000000000001* np.identity(len(m)), 10)
#plt.scatter(motorReg.featTest, motorReg.predict(motorReg.xTest))
for i in range(10):
    motorReg.w = np.reshape(w[i, :], (-1,1))
    #print(motorReg.w.shape)
    plt.plot(motorReg.featGrid, motorReg.predict(motorReg.xGrid))
plt.scatter(motorReg.featTest, motorReg.yTest, label='test data')
plt.legend()
plt.title('Random instances of prediction function for radial features')
plt.show()




```
